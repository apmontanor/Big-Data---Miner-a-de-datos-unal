**Ronald Aylmer Fisher**

<p align ="justify"> Las bases teóricas de estos modelos las dieron Gauss y Legendre a comienzos del siglo XIX, ellos describen el método de los mínimos cuadrados desarrollado bajo el supuesto de la normalidad de los errores. Sin embargo, la metodología basada en la máxima verosimilitud (MV) propuesta por Fisher en el primer cuarto del siglo XX, da credibilidad a la teoría de los mínimos cuadrados. Fisher presentó inicialmente el procedimiento de MV en 1912 y se considera uno de los desarrollos más importantes del siglo XX en la estadística.

<p align ="justify"> Fisher en 1912 propuso el “criterio absoluto” que fue derivado inicialmente del
principio “de la probabilidad inversa”. En 1921 desarrolló el principio del “óptimo” el cual se asociaba con la noción de “verosimilitud”. El método de MV propuesto en 1922 produce estimadores que satisfacen los criterios de suficiencia y eficiencia o cuando se hace MV se llega a un mundo de ideas y nomenclatura estadística, tales como parámetros, estadística, verosimilitud, suficiencia, consistencia, eficiencia e información. El método de MV constituye una herramienta importante en el desarrollo de estimación tanto en los modelos lineales clásicos como en la estructura de los modelos lineales generalizados (MLG) y los modelos de dispersión, y en general, de casi todas las áreas del conocimiento científico.

<p align ="justify"> En los libros clásicos tales como Draper & Smith (1966), Scheffe (1959) y Graybill (1961), se sintetizan los desarrollos teóricos de los modelos lineales alcanzados en la primera mitad del siglo XX. Existen dos clases generales de modelos que tienen como soporte teórico la distribución normal: los modelos de regresión y los modelos de análisis de varianza, cada uno de estos modelos pueden ser vistos como modelos super parametrizados y/o modelo de medias de celdas, introducido por Hocking & Speed (1975).

<p align ="justify"> La respuesta de una variable de interés conocida como variable de predicción, va a depender de unas variables conocidas como regresoras, predictoras, o variables independientes, típicamente estas variables son cuantitativas, pero también variables cualitativas pueden ser consideradas en el conjunto de las variables predictoras, esto lleva al estudio de modelos de regresión tipo Poisson, Gamma, Beta, etc., según se la naturaleza de los datos.

<p align ="justify"> Los modelos de regresión tienen cuatro tipos de uso fundamental: descripción y
aplicación, control, predicción y calibración. Brevemente se puede afirmar que:



*   <p align ="justify"> Los modelos de regresión son valiosos para describir el tipo de asociación entre la variable respuesta $y_i$ y las variables predictoras $(x_{i1}, x_{i2}, . . . , x_{ij} , . . . , x_{ip})$, $i = 1, . . . , n$. En este caso, lo que se persigue es resumir las tendencias de los datos y encontrar la forma de asociación entre las variables.
*Un uso importante de los modelos de regresión es que se tiene la posibilidad de predecir el valor que tendría $y_i$, o la media de la población del conjunto de $y_i$, $(E [Y_i])$ cuando se especifican las condiciones del proceso mediante los valores de $x_{ij}$ , $i = 1, . . . , n$, $j = 1, . . . , p$.

*   <p align ="justify"> Un modelo de regresión puede servir para encontrar cuáles son los valores de los $x_j (j = 1, . . . , p)$ que pueden optimizar, de acuerdo con algún criterio los valores de la variable dependiente yi. Este sistema de control es usado ampliamente en ensayos con fertilizantes, funciones de producción y curvas de crecimiento.

*   <p align ="justify"> La regresión es empleada también en problemas de calibración, en donde Yi
es una característica aleatoria, fácil de medir que depende de una variable no
aleatoria difícil de medir. En este caso, la relación de dependencia se puede
medir mediante una regresión del tipo $$y_i=\sum_{j=1}^{p}{\beta_j g_j(x_i)+e_i} \quad i=1,\dots, n.$$
Este modelo se usa para predecir el valor de $x_j$ que ha producido un valor
determinado para $y_i$.

**3. Planteamiento del algoritmo**

*Caso particular regresión lineal simple*


<p align ="justify"> La regresión lineal simple parte de una sóla variable predictora, es decir $X=x_1$ donde $X \in \mathcal{R}$ y supone que existe aproximadamente una relación lineal entre $X$ e $Y$. Esta relación lineal puede ser escrita como:

$$Y = \beta_0 + \beta_1 X + \varepsilon$$

<p align ="justify"> Donde $\beta_0 + \beta_1 X$ corresponde a la parte del modelo de regresión lineal, siendo $\beta_0$ y $\beta_1$ el intercepto y la pendiente respectivamente, y $\varepsilon$ corresponde al error cometido por el modelo.

<p align ="justify"> Los coeficientes $\beta_0$ y $\beta_1$ son dos constantes que se obtienen por medio de lo que se denomina *entrenamiento* del modelo de regresión. Este entrenamiento se realiza con datos de los que se sabe el valor real de la tupla $(X,y)$. Una vez obtenida la estimación de estos dos coeficientes se puede predecir la variable Y usando X conocidos y a partir de la siguiente fórmula:
$$\hat{y}=\hat{\beta}_0 +\hat{\beta}_1 x$$

<p align ="justify"> Siendo $\hat{y}$ la predicción obtenida de la variable $y$, y $\hat{\beta}_0$ y $\hat{\beta}_1$ son las estimaciones obtenidas con el entrenamiento.

**Estimación de los coeficientes**

<p align ="justify"> La estimación de los coeficientes se realiza con valores reales conocidos de la tupla $(X,y)$. Suponga que se tienen n registros con variable independiente y dependiente conocidas:

$$(x_1,y_1),(x_2,y_2),\ldots,(x_n,y_n)$$

<p align ="justify"> Con estas n observaciones, el objetivo es obtener las estimaciones de los coeficientes $\hat{\beta}_0$ y $\hat{\beta}_1$ que hagan que el modelo lineal se ajuste lo mejor posible a los datos. Es decir, que $\hat{y}_i \approx \hat{\beta}_0 +\hat{\beta}_1 x_i$ para todos los $x_i$, o lo que es lo mismo, encontrar la pendiente y la ordenada en el origen de la recta que pasa más cerca de los puntos $(x_i,y_i)$.

<p align ="justify"> Esta medida de cercania entre la recta y los puntos se puede hacer de distintas formas, siendo la forma más común la de ajustar por mínimos cuadrados (less squares criterion).

**Estimación por mínimos cuadrados**

<p align ="justify"> Siendo $\hat{y}_i \approx \hat{\beta}_0 +\hat{\beta}_1 x_i$ las predicciones de $y_i$, el residuo o diferencia entre la predicción y el valor real es:

$$r_i=y_i - \hat{y}_i$$

<p align ="justify"> El método de mínimos cuadrados obtendrá los coeficientes que minimicen la suma de residuos al cuadrado, es decir, que minimicen:

$$RSS=\sum_{i=1}^n r_i^2$$

<p align ="justify"> Usando derivadas parciales e igualando a 0 obtenemos que:

$$\frac{\partial RSS}{\partial \hat{\beta}_1} = 0 \rightarrow \hat{\beta}_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}$$

$$\frac{\partial RSS}{\partial \hat{\beta}_0} = 0 \rightarrow \hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$$

<p align ="justify"> Siendo $\bar{x}$ e $\bar{y}$ las medias aritméticas de las muestras.

**4. Ejemplo con datos reales**

```
# Importar librerías de trabajo
import numpy as np
import pandas as pd
import random
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Generador de distribución de datos para regresión lineal simple
def generador_datos_simple(beta, muestras, desviacion):
# Genero n (muestras) valores de x aleatorios entre 0 y 100
  x = np.random.random(muestras) * 100
# Genero un error aleatorio gaussiano con desviación típica (desviacion)
  e = np.random.randn(muestras) * desviacion
# Obtengo el y real como x*beta + error
  y = x * beta + e
  return x.reshape((muestras,1)), y.reshape((muestras,1))
```

```
# Parámetros de la distribución
desviacion = 200
beta = 10
n = 50
x, y = generador_datos_simple(beta, n, desviacion)
```

```
x1=pd.DataFrame(data=x)
y1=pd.DataFrame(data=y)
```

```
# Represento los datos generados
plt.scatter(x, y)
plt.show()
```

```
# Creo un modelo de regresión lineal
modelo = linear_model.LinearRegression()
```

```
# Entreno el modelo con los datos (X,Y)
modelo.fit(x, y)
# Ahora puedo obtener el coeficiente b_1
print('Coeficiente beta1: ', modelo.coef_[0])
```

```
# Podemos predecir usando el modelo
y_pred = modelo.predict(x)
```

```
# Por último, calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y, y_pred))
print('Estadístico R_2: %.2f' % r2_score(y, y_pred))
```

```
# Representamos el ajuste (rojo) y la recta Y = beta*x (verde)
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
x_real = np.array([0, 100])
y_real = x_real*beta
plt.plot(x_real, y_real, color='green')
plt.show()
```
